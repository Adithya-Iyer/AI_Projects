Task 4: Data Structure

```python
def task4():
  data = {'school': 'UAlbany', 'address': '1400 Washington Ave, Albany, NY 12222', 'phone': '(518) 442-3300'}
  vals = set(['UAlbany', '1400 Washington Ave, Albany, NY 12222', '(518) 442-3300'])
  print('school:', data.get('school', '--missing--'))
  if vals!=set(data.values()):
    return
  if 'address' in data.keys():
    print('address:', data['address'])
  for key, value in data.items():
    if key=='phone':
      print('phone:', value)
  
task4()
```

Task 5: Data Serialization

```python
import json

def task5():
  data = {'school': 'UAlbany', 'address': '1400 Washington Ave, Albany, NY 12222', 'phone': '(518) 442-3300'}
  f = open('data.json', 'w')
  json.dump(data, f)
  f.close()
  fp = open('data.json', 'r')
  data = json.load(fp)
  print(data)
  print(data.items())
  print(data.keys())
  print(data.values())

task5()
```

Task 6: Data Serialization

```python
import pickle
import json

def task6():
  data = {'school': 'UAlbany', 'address': '1400 Washington Ave, Albany, NY 12222', 'phone': '(518) 442-3300'}
  items = [1,2,3,4,5]
  
  f = open('data.json', 'w')
  json.dump(data, f)
  f.close()
  fp = open('data.json', 'r')
  data = json.load(fp)
  print(data)
  fp.close()

  f = open('file', 'wb')
  pickle.dump(items, f)
  f.close()
  fp = open('file', 'rb')
  items = pickle.load(fp)
  print(items)
  fp.close()

task6()
```

Task 8: Data Preprocessing: tweets filtering

```python
import json
from datetime import datetime

def task8():
  tweetList = []
  for line in open('CrimeReport.txt', 'r').readlines():
    tweet = json.loads(line)
    tweetList.append(tweet)
  recentTweets = sorted(tweetList, key=lambda x: datetime.strptime(x['created_at'], "%a %b %d %H:%M:%S +0000 %Y"))
  with open('task8.data', 'w') as f:
    for newTweet in recentTweets[-10:]:
      f.write(json.dumps(newTweet)+'\n')

task8()
```

Task 9: File operations

```python
import os
import json

def task9():
  os.chdir('/content') #making sure we are in the right directory to begin with
  if os.path.exists('task9-output')==False:
    os.mkdir('task9-output')
  os.chdir('task9-output')
  for line in open('/content/CrimeReport.txt', 'r').readlines():
    tweet = json.loads(line)
    timestamp = tweet['created_at'].split()
    filename = timestamp[1]+'-'+timestamp[2]+'-'+timestamp[-1]+'-'+timestamp[3][:2]+'.txt'
    with open(filename, 'a') as f:
      f.write(json.dumps(tweet)+'\n')

task9()
```
